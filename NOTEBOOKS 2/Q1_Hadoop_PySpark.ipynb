## Data Cleaning

Checking for missing values and duplicates before the analysis.

```python
from pyspark.sql.functions import col

# Check for missing values in the key columns
key_columns = ["country", "temperature_celsius", "humidity", "wind_kph", "condition_text", "air_quality_PM2.5"]

for c in key_columns:
    print(c, ":", df.filter(col(f"`{c}`").isNull()).count())
```
